###  Control Statements

- Control Statements are used to Control the flow of Execution of a Program.
- They decide,which statements to execute,how many times execute,and when to stop Execution.

#### Why Control Statements are Important

- Make Decisions
- repeat tasks automatically
- Control Program Execution
- reduce the code repetation

### Types of Control Statements

##### 1.Conditional Statements
##### 2.Looping Statements
##### 3.Jumping Statements

### 1.Conditional Statements

- Used check the conditions and  execution code based on the True,False

#### Why Do We Use Conditional Statements?
- To make decisions in a program.
- To execute code only when a condition is true.
- To choose between different actions.

```python
# example:

If it is raining
    Take an umbrella
Else
    Wear sunglasses
```

#### Types of Conditional Statements
- 1.if
- 2.if...else
- 3.if...elif...else
- 4.Nested if

### 3. if...elif...else Statement

- The elif (else if) statement is used to check multiple conditions.

```python
# Syntax
if condition1:
    statement
elif condition2:
    statement
elif condition3:
    statement
else:
    statement
```

```python
# Flow Diagram
Condition1
    │
 ┌──┴──┐
 │True │
 └──┬──┘
    │
 Execute
    │
 False
    ▼
Condition2
    │
 ┌──┴──┐
 │True │
 └──┬──┘
    │
 Execute
    │
 False
    ▼
Else Block
```

```python
marks=int(input("enter a marks:"))
if marks>90:
    print(f"Grade A")
elif marks>75:
    print(f"Grade B")
elif marks>60:
    print(f"Grade C")
elif marks>45:
    print(f"Grade D")
else:
    print("Fail")
```

### 4. Nested if

- A Nested if means placing one if statement inside another if statement.

```python
# syntax
if condition1:
    if condition2:
        statement
```

```python
age = 22
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to vote")
```

```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login Successful")
```

### Comparison Operators Used in Conditions

| Operator | Meaning                  |
| -------- | ------------------------ |
| `==`     | Equal to                 |
| `!=`     | Not Equal to             |
| `>`      | Greater Than             |
| `<`      | Less Than                |
| `>=`     | Greater Than or Equal To |
| `<=`     | Less Than or Equal To    |


### Examples with Logical Operators

**and**

```python
age=20
citizen=True

if age>=18 and citizen:
    print("eligible for vote")
```

**or**

```python
username = "admin"

if username == "admin" or username == "manager":
    print("Access Granted")
```

**not**

```python
is_logged_in = False

if not is_logged_in:
    print("Please Login")
```

#### Summary Table

| Statement          | Purpose                                |
| ------------------ | -------------------------------------- |
| `if`               | Executes code if the condition is True |
| `if...else`        | Chooses between two blocks             |
| `if...elif...else` | Checks multiple conditions             |
| Nested `if`        | Places one `if` inside another         |


### Difference Between if, if...else, and elif

| Statement          | Use Case              |
| ------------------ | --------------------- |
| `if`               | Single condition      |
| `if...else`        | Two possible outcomes |
| `if...elif...else` | Multiple conditions   |


#### Key Points to Remember
- if is used to execute code only when a condition is True.
- if...else is used when there are two possible outcomes.
- if...elif...else is used to check multiple conditions.
- **Nested if** means placing one if statement inside another.
- Python uses indentation (typically 4 spaces) to define code blocks.
- Conditional statements often use comparison and logical operators to evaluate conditions.
