## Data Types in Python

### What is a Data Type?

##### Definition

- A data type tells Python what kind of data a variable stores.

- Simple Definition

  - A data type is the type or category of data stored in a variable.

### Why Do We Need Data Types?

- Different types of data require different operations.

- For example:

  - Age → Number (int)
  - Salary → Decimal Number (float)
  - Name → Text (str)
  - Student Status → True/False (bool)

- Python uses data types to understand how to store and process data.

### Types of Data Types in Python

- Python has the following built-in data types:

- int 
- float
- complex 
- bool 
- str 
- list 
- tuple 
- set 
- dictionary 
- None

### 1. Integer (int)
- Definition

  - An integer is a whole number without a decimal point.

```python
# example:
age=21
marks=95
temp=-5
```

```python
print(type(age))
print(type(marks))
print(type(temp))
```

#### Characteristics
- Positive numbers
- Negative numbers
- Zero
- No decimal point

### 2. Float (float)
- Definition

  - A float is a number that contains a decimal point.

```python
# example:
salary=20000.0
height=5.6
pi=3.14
```

```python
print(type(salary))
print(type(height))
print(type(pi))
```

#### Characteristics
- Contains decimal values
- Can be positive or negative

### 3. Complex (complex)
- Definition

- A complex number has two parts:

   - Real part
   - Imaginary part

- The imaginary part is written using j.

```python
# example:
x=3+4j
```

```python
print(type(x))
```

#### Note

- Complex numbers are mainly used in scientific and engineering calculations.

### 4. Boolean (bool)
- Definition

- A Boolean data type has only two values:

   - True
   - False

```python
is_student=True
is_logged_in=False
```

```python
print(type(is_student))
print(type(is_logged_in))
```

#### Uses
- Decision making
- Conditions
- Login status
- Yes/No questions

### 5. String (str)
- Definition

- A string is a sequence of characters enclosed in quotes.

```python
name="Ramya"
city="Vizag"
```

```python
print(type(name))
print(type(city))
```

#### String can contain
 - Letters
 - Numbers
 - Symbols
 - Spaces

```python
address="Hyderabad 500001"
```

```python
print(type(address))
```

```python
name='ramya'
type(name)
```

```python
name="""Ramya"""
print(type(name))
```

### 6. List (list)
- Definition

  - A list is an ordered collection of items.

  - Lists are mutable, which means you can change their contents after creation.

```python
num=[1,2,3,4,5]
```

```python
print(type(num))
```

#### Characteristics
- Ordered
- Mutable (can be modified)
- Allows duplicate values
- Uses square brackets []

### 7. Tuple (tuple)
- Definition

  - A tuple is an ordered collection of items.

  - Tuples are immutable, which means they cannot be changed after creation.

```python
numbers = (10, 20, 30, 40)
```

```python
print(type(numbers))
```

#### Characteristics
- Ordered
- Immutable
- Allows duplicates
- Uses parentheses ()

### 8. Set (set)
- Definition
 - A set is an unordered collection of unique items.

```python
num={10,20,30,40,40}
```

```python
print(type(num))
```

#### Characteristics
- Unordered
- No duplicate values
- Mutable
- Uses curly braces {}
- Duplicate values are automatically removed.

```python
data = {10, 20, 20, 30}
print(data)
```

### 9. Dictionary (dict)
- Definition
- A dictionary stores data as key-value pairs.

```python
student={
    "name":"Ramya",
    "age":21,
    "ciyt":"vizag"
}
```

```python
type(student)
```

#### Characteristics
- Stores key-value pairs
- Mutable
- Uses curly braces {}

### 10. None (NoneType)
- Definition
- None represents no value or nothing.

```python
x = None
```

```python
print(type(x))
```

#### Uses
- Placeholder for a value that will be assigned later
- Represents the absence of a value
