## Dictionaries in Python

### 1. Introduction
**What is a Dictionary?**

- A **Dictionary is a collection of key-value pairs**.

   - Each key is unique.
   - Keys are used to access values.
   - Dictionaries are **mutable** (modifiable).
   - Dictionaries are enclosed in **curly braces {}**.

```python
# Syntax
dictionary_name = {
    key1: value1,
    key2: value2
}
```

```python
# example
student={
    "name":"ramya",
    "age":21,
    "city":"Vizag"
    }
student
```

### Why Dictionary?

- Dictionaries are useful when:
   - Data has a key-value relationship.
   - Fast searching using keys.
   - Easy to update values.
   - Efficient for storing structured data.
    
### Features of Dictionaries
   - Stores key-value pairs
   - Mutable
   - Keys are unique
   - Values can be duplicated
   - Ordered (Python 3.7+)
   - Fast lookup using keys

```python
# example
employee={
    "id":101,
    "name":"Shanthi",
    "salary":50000
    }
employee
```

### 3. Accessing Elements
#### Using Keys

```python
student = {
    "name":"Ramya",
    "age":22
}
print(student["name"])
print(student["age"])
```

#### values()

```python
student = {
    "name":"Ramya",
    "age":22
}
print(student.values())
```

#### get()
- Returns the value without raising an error if the key doesn't exist.

```python
student = {
    "name":"Ramya",
    "age":22
}
print(student.get("name"))
print(student.get("city"))
```

##### Using a default value:

```python
print(student.get("city", "Not Found"))
```

### 4. Dictionary Methods
#### keys()

- Returns all keys.

```python
student = {
    "name":"Ramya",
    "age":22
}
print(student.keys())
```

#### values()

- Returns all values.

```python
print(student.values())
```

#### items()
- Returns key-value pairs.

```python
print(student.items())
```

#### update()

- Updates or adds key-value pairs.

```python
student = {
    "name":"Ramya",
    "age":22
}
student.update({"city":"Hyderabad"})
print(student)
```

#### pop()

- Removes a specified key.

```python
student = {
    "name":"Ramya",
    "age":22
}
student.pop("age")
print(student)
```

#### popitem()

- Removes the last inserted key-value pair.

```python
student = {
    "name":"Ramya",
    "age":22
}
student.popitem()
print(student)
```

#### clear()
- Removes all items.

```python
student = {
    "name":"Ramya",
    "age":22
}
student.clear()
print(student)
```

#### copy()

- Creates a copy.

```python
student = {
    "name":"Ramya",
    "age":22
}
new_student = student.copy()
print(new_student)
```

#### setdefault()

- Returns the value of a key. If the key doesn't exist, it inserts the key with a default value.

```python
student = {
    "name":"Ramya"
}
student.setdefault("age",22)
print(student)
```

#### fromkeys()

- Creates a new dictionary from given keys.

```python
keys = ("name","age","city")
d = dict.fromkeys(keys)
print(d)
```

##### With a default value

```python
d = dict.fromkeys(keys,0)
print(d)
```

### 5. Dictionary Operations
#### Add a New Key

```python
student = {
    "name":"Ramya"
}
student["age"] = 22
print(student)
```

#### Modify a Value

```python
student = {
    "name":"Ramya",
    "age":22
}
student["age"] = 23
print(student)
```

#### Delete a Key

```python
student = {
    "name":"Ramya",
    "age":22
}
del student["age"]
print(student)
```

#### Iterate Through a Dictionary
**Iterate Keys**

```python
student = {
    "name":"Ramya",
    "age":22
}
for key in student:
    print(key)
```

**Iterate Values**

```python
for value in student.values():
    print(value)
```

**Iterate Key-Value Pairs**

```python
for key, value in student.items():
    print(key, ":", value)
```

### 6. Nested Dictionary
- Creating a Nested Dictionary

```python
students = {
    "student1":{
        "name":"Ramya",
        "age":22
    },
    "student2":{
        "name":"Ravi",
        "age":23
    }
}
print(students)
```

### Accessing Nested Dictionary

```python
print(students["student1"]["name"])
print(students["student2"]["age"])
```

### 7. Dictionary Comprehension
- Basic Dictionary Comprehension

```python
square = {x:x*x for x in range(1,6)}
print(square)
```

#### Dictionary Comprehension with Condition

```python
even = {x:x*x for x in range(1,11) if x%2==0}
print(even)
```

### 8. Common Dictionary Programs
#### Count Frequency of Elements 

```python
text = "apple"
freq = {}
for ch in text:
    freq[ch] = freq.get(ch,0) + 1
print(freq)
```

#### Merge Two Dictionaries

```python
d1 = {"a":1,"b":2}
d2 = {"c":3,"d":4}
d1.update(d2)
print(d1)
```

#### Invert a Dictionary

```python
student = {
    "a":1,
    "b":2,
    "c":3
}
inverse = {v:k for k,v in student.items()}
print(inverse)
```

#### Sort Dictionary by Keys

```python
student = {
    "c":3,
    "a":1,
    "b":2
}
for key in sorted(student):
    print(key, student[key])
```

#### Find Maximum Value

```python
marks = {
    "Python":95,
    "SQL":90,
    "AI":98
}
print(max(marks.values()))
```

#### Find Minimum Value

```python
marks = {
    "Python":95,
    "SQL":90,
    "AI":98
}
print(min(marks.values()))
```

### Difference Between List, Tuple, Set, and Dictionary

| List              | Tuple             | Set           | Dictionary    |
| ----------------- | ----------------- | ------------- | ------------- |
| `[]`              | `()`              | `{}`          | `{key:value}` |
| Ordered           | Ordered           | Unordered     | Ordered       |
| Mutable           | Immutable         | Mutable       | Mutable       |
| Allows duplicates | Allows duplicates | No duplicates | Unique keys   |
| Indexed           | Indexed           | No indexing   | Access by key |


⭐ **Summary**
- A **dictionary stores data as key-value pairs**.
- Dictionaries are **mutable, ordered** (Python 3.7+), and require **unique keys**.
- Create dictionaries using {} or the dict() constructor.
- Access values using **keys**, [], or the safer get() method.
- Common methods include keys(), values(), items(), update(), pop(), popitem(), clear(), copy(), setdefault(), and fromkeys().
- Dictionaries support adding, modifying, deleting, and iterating through key-value pairs.
- **Nested dictionaries** allow you to store complex, hierarchical data.
- Dictionary comprehensions provide a concise way to create dictionaries.
- Common applications include **counting frequencies, merging dictionaries, inverting key-value pairs, sorting**, and finding the **maximum or minimum values**.
