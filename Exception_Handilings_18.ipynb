## Exception Handling in Python

### 1. Introduction
**What is Exception Handling?**

- **Exception Handling** is the process of handling runtime errors so that the program does not terminate unexpectedly.
- Instead of crashing, the program handles the error gracefully and continues execution.

### Why Use Exception Handling?
- Prevents program crashes
- Handles runtime errors
- Improves program reliability
- Displays user-friendly error messages
- Allows the program to continue execution

| Exception             | Description                     | Example                           |
| --------------------- | ------------------------------- | --------------------------------- |
| `SyntaxError`         | Invalid Python syntax           | Missing `:`                       |
| `IndentationError`    | Incorrect indentation           | Missing spaces/tabs               |
| `NameError`           | Variable not defined            | `print(x)` when `x` doesn't exist |
| `TypeError`           | Invalid operation between types | `"10" + 5`                        |
| `ValueError`          | Correct type but invalid value  | `int("abc")`                      |
| `ZeroDivisionError`   | Division by zero                | `10 / 0`                          |
| `IndexError`          | List index out of range         | `a[5]`                            |
| `KeyError`            | Dictionary key not found        | `d['age']`                        |
| `AttributeError`      | Object has no attribute         | `"abc".append(1)`                 |
| `ImportError`         | Module cannot be imported       | `import xyz`                      |
| `ModuleNotFoundError` | Module does not exist           | `import mymodule`                 |
| `FileNotFoundError`   | File does not exist             | `open("abc.txt")`                 |
| `EOFError`            | No input received               | `input()` at end of file          |
| `OverflowError`       | Number too large                | Very large calculations           |
| `MemoryError`         | Out of memory                   | Creating extremely large objects  |
| `RecursionError`      | Maximum recursion exceeded      | Infinite recursive function       |
| `AssertionError`      | Assertion failed                | `assert x > 0`                    |
| `RuntimeError`        | General runtime error           | Program execution issues          |
| `KeyboardInterrupt`   | User interrupts program         | Pressing `Ctrl + C`               |
| `StopIteration`       | Iterator has no more values     | `next()` on exhausted iterator    |

#### 2. try Block
- The try block contains code that may generate an exception.

```python
# syntax
try:
    # risky code
```

```python
#example
try:
    a=10
    b=20
    print(a/b)
except:
    print("error")
```

#### 3. except Block
- The except block executes only if an exception occurs

```python
# syntax
try:
    # risky code
except:
    # handle  exception   
```

```python
#example
try:
    print(10/0)
except:
    print("cannot divide by zero")
```

**Specific Exception**

```python
try:
    print(10/0)
except ZeroDivisionError:
    print("Division by zero is not allowed.")
```

### Multiple Exceptions

```python
try:
    num=int(input("enter a number:"))
    print(10/num)
except ValueError:
    print("please enter a valid number.")
except ZeroDivisionError:
    print("cannot divide by zero.")
```

#### 4. else Block
- The else block executes only when no exception occurs.

```python
# Syntax
try:
    # risky code
except:
    # error handling
else:
    # executes if no error
```

```python
# example
try:
    a=10
    b=20
    print(a/b)
except ZeroDivisionError:
    print("division error")
else:
    print("division successfully")
```

#### 5. finally Block

- The finally block always executes whether an exception occurs or not.

- It is mainly used for:

   - Closing files
   - Closing database connections
   - Releasing resources

```python
# syntax
try:
    ...
except:
    ...
finally:
    ...
```

```python
# Example
try:
    print(10/0)
except ZeroDivisionError:
    print("cannot divide by zero")

finally:
    print("program finished")
```

```python
# Example (No Exception)
try:
    print(10/2)

except ZeroDivisionError:
    print('error')
finally:
    print("always executes")
```

#### 6. raise Statement
- The raise keyword is used to manually generate an exception.

```python
# syntax
raise ExceptionType("Message")
```

```python
# example
age=15
if age<18:
    raise ValueError("age must be at least 18.")
```

```python
# example
num=-5
if num<0:
    raise Exception("negative numbers are not allowed")
```

#### 7. Exception Object

```python
try:
    print(10/0)

except Exception as e:
    print("error:",e)
# e stores the exception message.
```

#### 8. Complete Example

```python
try:
    a=int(input("enter a first value:"))
    b=int(input("enter a second value:"))

    result=a/b

except ValueError:
    print("please enter valid number.")

except ZeroDivisionError:
    print("cannot divide by zero.")
else:
    print("Result=",result)
finally:
    print("Program ended")
```

#### 9. Nested try-except

```python
try:
    try:
        print(10/0)
    except ZeroDivisionError:
        print("inner exception")

except:
    print("outer object")
```

#### 10. User-Defined Exception Using raise

```python
salary=3000
if salary<5000:
    raise Exception("salary must be at least 5000.")
```

#### 11. Flow of Exception Handling

```python
Start
   │
   ▼
try Block
   │
   ├── No Exception ──► else Block
   │
   └── Exception
          │
          ▼
      except Block
          │
          ▼
     finally Block
          │
          ▼
         End
```

**12. Advantages**
  -  Prevents program termination
  -  Improves user experience
  -  Makes debugging easier
  -  Handles runtime errors safely
  -  Ensures cleanup using finally
      
**13. Disadvantages**
- Too many try-except blocks can reduce readability.
- Catching every exception (except:) may hide real bugs.
- Slight performance overhead.

#### Difference Between Keywords

| Keyword   | Purpose                                   |
| --------- | ----------------------------------------- |
| `try`     | Contains code that may raise an exception |
| `except`  | Handles the exception                     |
| `else`    | Executes if no exception occurs           |
| `finally` | Always executes                           |
| `raise`   | Manually raises an exception              |
