## File Handling in Python

### 1. Introduction
- **What is File Handling?**

- **File Handling** is the process of creating, reading, writing, updating, and deleting files using Python.

- Python provides built-in functions to work with files easily.

### Why Use File Handling?
Store data permanently.
   - Read data from files.
   - Write program output to files.
   - Process CSV and JSON files.
   - Exchange data between applications.
       
### Types of Files
- **1. Text Files (.txt)**

- Stores data as plain text.

```python
#example
Hello Python
Welcome to File Handling
```

- **2. Binary Files (.jpg, .pdf, .mp3, etc.)**

- Stores data in binary format.

- Examples:

   - Image
   - Video
   - PDF
   - Audio

### 2. open() Function

- The open() function is used to open a file.

```python
#syntax
file = open("filename", "mode")
```

**File Modes**

| Mode   | Description                              |
| ------ | ---------------------------------------- |
| `"r"`  | Read (default)                           |
| `"w"`  | Write (Creates file if it doesn't exist) |
| `"a"`  | Append                                   |
| `"x"`  | Create a new file                        |
| `"r+"` | Read and Write                           |
| `"w+"` | Write and Read                           |
| `"a+"` | Append and Read                          |
| `"rb"` | Read Binary                              |
| `"wb"` | Write Binary                             |


### 3. Reading Files
**read()**
- Reads the entire file.

```python
# example
file=open("data.txt","r")
result=file.read()
print(result)
file.close
```

**readline()**
- Reads only one line.

```python
file=open("data.txt","r")
print(file.readline())
file.close()
```

**readlines()**
- Reads all lines and returns a list.

```python
file=open("data.txt","r")
print(file.readlines())
file.close()
```

##### Example
- Suppose sample.txt

```python
Python
SQL
Machine Learning
```

```python
file=open("sample.txt","r")
print(file.read())
```

```python
Python
SQL
Machine Learning
```

### 4. Writing Files
- Write mode **overwrites** existing content.

```python
# examples
file=open("sample.txt","r")
file.write("hello python")
file.close()
```

#### Write Multiple Lines

```python
file = open("sample.txt", "w")

file.write("Python\n")
file.write("SQL\n")
file.write("AI\n")

file.close()
```

### 5. Append Files

- Append mode adds data without deleting existing content.

```python
# Example
file=open("sample.txt","r")
file.write("\nMachine Learning")
file.close()
```

```python
Suppose file already contains:
    
Python
SQL

After append:-

Python
SQL
Machine Learning
```

### 6. Closing Files

- Always close files after use.

```python
file.close()
# This releases system resources
```

### 7. with Statement
- The with statement automatically closes the file.

```python
# syntax
with open("simple.txt","r") as file:
    print(file.read())
```

### 9. CSV Files
**What is CSV?**

CSV means

**Comma Separated Values**

```python
# example
Name,Age,City
Ramya,22,Hyderabad
Ravi,23,Vijayawada

# python uses the CSV modules
```

#### Writing CSV

```python
import csv

with open("students.csv","w",newline="") as file:

    writer = csv.writer(file)

    writer.writerow(["Name","Age","City"])

    writer.writerow(["Ramya",22,"Hyderabad"])

    writer.writerow(["Ravi",23,"Vijayawada"])
```

#### Reading CSV

```python
import csv
with open("students.csv","r") as file:

    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
# output
['Name', 'Age', 'City']
['Ramya', '22', 'Hyderabad']
['Ravi', '23', 'Vijayawada']
```

### 10. JSON Files
#### What is JSON?

JSON stands for

**JavaScript Object Notation**

- It stores data in key-value format.

```python
# example
{
    "name":"Ramya",
    "age":22,
    "city":"Hyderabad"
}

# python uses the json module
```

#### Convert Dictionary to JSON

```python
import json
student = {
    "name":"Ramya",
    "age":22,
    "city":"Hyderabad"
}
with open("student.json","w") as file:

    json.dump(student,file,indent=4)
```

```python
# Generated JSON
{
    "name": "Ramya",
    "age": 22,
    "city": "Hyderabad"
}
```

#### Read JSON File

```python
import json
with open("student.json","r") as file:

    data = json.load(file)

print(data)
```

```python
#output
{'name': 'Ramya', 'age': 22, 'city': 'Hyderabad'}
```

### 11. Common File Programs
#### Program 1: Read a File

```python
with open("sample.txt","r") as file:
    print(file.read())
```

#### Program 2: Write to a File

```python
with open("sample.txt","w") as file:
    file.write("welcome to python")
```

#### Program 3: Append Data

```python
with open("sample.txt","a") as file: 
    file.write("\nFile Handling")
```

#### Program 4: Count Number of Lines

```python
with open("sample.txt","r") as file:
    lines = file.readlines()

print("Lines =", len(lines))
```

#### Program 5: Count Words

```python
with open("sample.txt","r") as file:
    words = file.read().split()

print("Words =", len(words))
```

### Program 6: Count Characters

```python
with open("sample.txt","r") as file:
    text = file.read()

print("Characters =", len(text))
```

### Program 7: Copy File

```python
with open("sample.txt","r") as source:
    data = source.read()

with open("copy.txt","w") as destination:
    destination.write(data)
```

### Program 8: Read CSV File

```python
import csv
with open("students.csv","r") as file:

    reader = csv.reader(file)

    for row in reader:
        print(row)
```

### Program 9: Read JSON File

```python
import json
with open("student.json","r") as file:

    data = json.load(file)

print(data)
```

### Difference Between File Modes
| Mode | Purpose           | File Exists?          |
| ---- | ----------------- | --------------------- |
| `r`  | Read only         | Must exist            |
| `w`  | Write (overwrite) | Creates if not exists |
| `a`  | Append            | Creates if not exists |
| `x`  | Create new file   | Error if file exists  |
| `r+` | Read and Write    | Must exist            |
| `w+` | Write and Read    | Creates if not exists |
| `a+` | Append and Read   | Creates if not exists |


### Difference Between CSV and JSON
| CSV                    | JSON                     |
| ---------------------- | ------------------------ |
| Comma-separated values | Key-value pairs          |
| Tabular data           | Hierarchical/Nested data |
| Less flexible          | More flexible            |
| Uses `csv` module      | Uses `json` module       |
| Smaller file size      | Slightly larger          |
