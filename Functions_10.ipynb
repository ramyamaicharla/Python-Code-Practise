#### What is a Function?

- A function is a block of code that performs a specific task and can be called whenever needed.

**Simple Definition**

### Why Do We Use Functions?
- Avoid writing the same code multiple times.
- Improve code readability.
- Make programs easier to maintain.
- Divide a large program into smaller parts.
- Reuse code whenever needed.

### 4. Recursive Function
- A function that calls itself is called a recursive function.

- Simple Definition

  - A recursive function is a function that calls itself until a stopping condition is reached.

```python
# Syntax
def function():
    function()
```

#### Example

#### Factorial

```python
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5))
```

```python
# How Recursion Works
factorial(5)
│
├── 5 × factorial(4)
│
├── 4 × factorial(3)
│
├── 3 × factorial(2)
│
├── 2 × factorial(1)
│
└── 1
```

#### Base Case
- Every recursive function must have a base case.

### Using global

```python
count = 10

def update():
    global count
    count = 20

update()

print(count)
```

### Global and Local variables

```python
# Local variables
def display():
    x=10
    print("inside function:",x)
display()                                            
```

```python
def NAME():
    name="sravani"
    print("name:",name)
NAME()
```

```python
def details():
    salary=50000
    print("salary:",salary)
details()  
```

```python
# Global variables
comapny="TCS"
def cmpy():
    print("comapny:",comapny)
cmpy()
```

```python
x = 100      # Global variable
def display():
    print("Inside function:", x)
    
display()
print("Outside function:", x)
```

```python
# 3. Local and Global Variables with the Same Name

x = 100   # global variable

def display():
    x = 50     # Local variable
    print("Inside:", x)

display()

print("Outside:", x)
```

```python
country="india"
def COUNTRY():
    country="pakistan"
    print("country        :",country)
COUNTRY()
print("outside country:",country)
```

```python
# 4. Modifying a Global Variable Using global

x = 100
def display():
    global x
    x = 200
    print("Inside:", x)

display()
print("Outside:", x)
```

```python
#5. Without global

x = 100
def display():
    x = 200
    print("Inside:", x)

display()
print("Outside:", x)
```

| Local Variable                            | Global Variable                          |
| ----------------------------------------- | ---------------------------------------- |
| Declared inside a function                | Declared outside all functions           |
| Accessible only inside that function      | Accessible throughout the program        |
| Exists only while the function is running | Exists until the program ends            |
| Cannot be used outside the function       | Can be used inside and outside functions |
