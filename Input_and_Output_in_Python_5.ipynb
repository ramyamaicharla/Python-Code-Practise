### Input and Output in Python
- Input and Output (I/O) are the basic ways a program interacts with the user.

  - Input → Data given by the user to the program.
  - Output → Information displayed by the program to the user.

#### What is Input?
- Definition

- Input is the data entered by the user into a program.

- Simple Definition

  - Input is the information given by the user to the computer.

```python
# Real-Life Example
# ATM Machine
Enter PIN
# Your PIN is the input.
```

### What is Output?
- Definition

- Output is the result produced by the program after processing the input.

- Simple Definition

  - Output is the result displayed by the computer after processing the input.

```python
Input:
10 + 20

Output:
30
```

### 1. input() Function
- Definition

- The input() function is used to take input from the user.

```python
# Syntax
variable=input("message")
```

```python
name=input("enter a name:")
print(name)
```

```python
name=input("enter a city:")
print(name)
```

### Important Point

- The input() function **always returns a string.**

```python
age=input("enter a age:")
print(type(age))
```

### Taking Integer Input
- use int()

```python
age=int(input("enter a age:"))
print(age)
```

### Taking Float Input

```python
salary=float(input("enter a salary:"))
print(salary)
```

### Taking Boolean Input

- Since input() returns a string, convert it manually if needed.

```python
status=input("are you a student? (True/False):")
status=status=="True"
print(status)
```

### 3. Output Formatting

- Output formatting makes the output look clean and readable.

- Python provides three common ways.

### Method 1: Using Comma

```python
name="ramya"
age=21

print("name:",name)
print("age:",age)
```

### Method 2: Using format()

- Definition

- The format() method inserts values into placeholders ({}).

```python
# Syntax
"{}".format(value)
```

```python
name="ramya"
print("my name is:{}".format(name))
```

#### Multiple Variables

```python
name = "Ramya"
age = 21

print("My name is {} and I am {} years old.".format(name, age))
```

#### Positional Index

```python
print("{1} {0}".format("Python", "Learn"))
```

#### Named Arguments

```python
print("Name: {n}, Age: {a}".format(n="Ramya", a=21))
```

#### 4. f-string (Formatted String Literal)

- Introduced in Python 3.6.

- It is the easiest and most recommended way to format strings.

```python
# Syntax
f"Text {variable}"
```

```python
name = "Ramya"
print(f"My name is {name}")
```

#### Multiple Variables

```python
name = "Ramya"
age = 21

print(f"My name is {name}. I am {age} years old.")
```

#### Expressions in f-strings

```python
a = 10
b = 20

print(f"Sum = {a+b}")
```

### Difference Between format() and f-string

| `format()`                     | f-string                      |
| ------------------------------ | ----------------------------- |
| Introduced before Python 3.6   | Introduced in Python 3.6      |
| Slightly longer syntax         | Short and easy to read        |
| Uses `.format()`               | Uses `f` before the string    |
| Good for older Python versions | Recommended for modern Python |


```python
name=input("enter a name:")
age=int(input("enter a age:"))
city=input("enter a city:")
price=round(float(input("enter a price:")))

print("\nStudent Details")
print('----------------------------')
print(f"name:{name}")
print(f"age:{age}")
print(f"city:{city}")
print(f"price:{price}")
```
