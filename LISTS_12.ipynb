### Lists in Python

#### 1. Introduction
##### What is a List?

- A List is a collection of multiple values stored in a single variable.

  - Lists can store different data types.
  - Lists are ordered.
  - Lists are mutable (can be modified).

```python
# syntax
list_name=[value1,value2,value3]
```

```python
numbers=[1,2,3,4,5]
print(numbers)
```

#### Why Lists?
- Instead of creating many variables,

```python
# Without List
a=10
b=20
c=30
d=40
```

```python
# Using List
numbers=[10,20,30,40]
```

### 6. List Copying

**Assignment**
- Both variables refer to the same list.

```python
a=[10,20]
b=a
b.append(30)
print(a)
print(b)
```

**Shallow Copy**

```python
a=[10,20,30]
b=a.copy()
b.append(40)
print(a)
print(b)
```

#### Deep Copy
- Deep copy copies all nested objects independently.

```python
import copy
a=[[1,2],[3,4]]
b=copy.deepcopy(a)
b[0][0]=100
print(a)
print(b)
```

### 9. Common List Programs

#### Largest Element

```python
nums=[10,20,30,40,90]
print(max(nums))
```

#### Smallest Element

```python
nums=[10,20,30,40,90]
print(min(nums))
```

#### Second Largest

#### Remove Duplicates

```python
nums=[10,20,30,40,90,10,20,10,30]
b=list(set(nums))
print(b)
```

#### Merge Lists

```python
a=[1,2]
b=[3,4]
print(a+b)
```

#### Rotate List
- left rotation

```python
numbers = [1,2,3,4,5]
numbers = numbers[1:] + numbers[:1]
print(numbers)
```

#### Reverse List

```python
nums=[1,2,3]
print(nums[::-1])
```

#### Count Even and Odd

```python
nums=[1,2,3,4,5,6,7,8,9]
ec=0
oc=0

for i in nums:
    if i%2==0:
        ec=ec+1
    else:
        oc=oc+1
       
print("even:",ec)
print("odds:",oc)        
```

#### Find Average

```python
nums=[1,2,3,4,5,6,7,8,9]
avg=sum(nums)/len(nums)
print(avg)
```

#### Sort List

```python
nums=[1,2,3,4,5,6,7,8,9]
sorted(nums)
```

#### ⭐ Summary
- **List** is an ordered, mutable collection.
- Supports **indexing, negative indexing, and slicing.**
- Common methods include append(), extend(), insert(), remove(), pop(), sort(), reverse(), and copy().
- Built-in functions like len(), sum(), max(), min(), and sorted() make list operations easier.
- Learn the difference between **assignment, shallow copy, and deep copy**.
- **List comprehensions** provide a concise way to create and transform lists.
- Nested lists are useful for representing **matrices** and multidimensional data.
- Master common interview programs such as finding the largest element, removing duplicates, reversing, rotating, merging, sorting, searching, and computing sums and averages.
