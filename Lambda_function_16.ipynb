### Lambda Functions in Python

### 1. Introduction
**What is a Lambda Function?**

- A **Lambda Function is a small anonymous function** (a function without a name).

  - It is defined using the lambda keyword.
  - It can take **any number of arguments.**
  - It can have **only one expression**.
  - The value of that expression is automatically returned.

```python
# Syntax
lambda arguments: expression
```

#### Why Use Lambda Functions?
- To write short functions in one line.
- Used with functions like map(), filter(), and sorted().
- Makes the code concise and readable.
- Useful for temporary functions.

### Features of Lambda Functions
- Anonymous (no function name)
- One-line function
- Can accept multiple arguments
- Contains only one expression
- Automatically returns the result

#### 2. Lambda vs Normal Function
**Normal Function**

```python
def add(a,b):
    return a+b
print(add(10,20))
```

**Lambda Function**

```python
add=lambda a,b:a+b
print(add(10,20))
```

#### 3. Lambda Examples
**Example 1: Square of a Number**

```python
square=lambda x:x*x
print(square(5))
```

**Example 2: Cube of a Number**

```python
cube=lambda x:x**3
print(cube(3))
```

**Example 3: Addition**

```python
add=lambda a,b:a+b
print(add(10,20))
```

**Example 4: Maximum of Two Numbers**

```python
maximum=lambda a,b:a if a>b else b
print(maximum(15,8))
```

**Example 5: Check Even or Odd**

```python
even=lambda x:"even" if x%2==0 else "odd"
print(even(12))
print(even(15))
```

#### 4. Lambda with map()
**What is map()?**

- map() applies a function to every element in an iterable.

```python
# Syntax
map(function, iterable)
```

```python
# example
nums=[10,20,30,40,50]
square=list(map(lambda x:x*x,nums))
print(square)
```

#### 5. Lambda with filter()

**What is filter()?**

- filter() selects elements that satisfy a condition.

```python
# syntax
filter(function, iterable)
```

```python
# example
nums=[1,2,3,4,5,6,7,8,9]
even=list(filter(lambda x:x%2==0,nums))
print(even)
```

#### 6. Lambda with sorted()
- **Sort by String Length**

```python
names = ["Ramya", "AI", "Python", "SQL"]
result=sorted(names,key=lambda x:len(x))
print(result)
```

#### Sort Dictionary by Values

```python
students = {
    "Ramya":95,
    "Ravi":85,
    "Sita":90
}
result=sorted(students.items(),key=lambda x:x[1])
print(result)
```

### 7. Lambda with Multiple Arguments

#### 8. Lambda with Conditional Expression

```python
result = lambda x: "Positive" if x > 0 else "Negative"
print(result(10))
print(result(-5))
```

### 10. Advantages of Lambda
- Short and simple syntax
- Reduces code size
- Useful with map(), filter(), sorted()
- No need to define a separate function

#### 11. Limitations of Lambda
- Can contain only one expression
- Cannot contain multiple statements
- Not suitable for complex logic
- Less readable for long expressions

# Difference Between def and lambda
| `def` Function              | `lambda` Function            |
| --------------------------- | ---------------------------- |
| Has a function name         | Anonymous (no name)          |
| Multiple statements allowed | Only one expression          |
| Uses `return` keyword       | Returns automatically        |
| Suitable for complex logic  | Suitable for short logic     |
| Easier to debug             | Best for temporary functions |


#### ⭐ Summary
- A **lambda function is an anonymous, one-line function** created using the lambda keyword.
- Syntax: lambda arguments: expression
- It automatically returns the result of the expression.
- Lambda functions are commonly used with:
- map() → Transform each element
- filter() → Select elements based on a condition
- sorted() → Customize sorting
- Lambda functions are ideal **for short, simple operations**, while def functions are better for **complex logic**.
