### List Comprehension in Python

### 1. Introduction
**What is List Comprehension?**

- list comprehension is used to write list operations in a single line
- the concept is save the results in a list
- we already know that append method is used to save the results
- similar to list comprehension also used to save results but in a single line
- Instead of writing a for loop and using append(), list comprehension allows you to create the list directly

#### Why Use List Comprehension?
- Reduces code length
- Improves readability
- Faster than traditional for loops
- Creates a new list in one line

#### Features
- Creates a new list
- Uses a single line of code
- Supports if and if-else conditions
- Can use nested loops
- Easy to read and maintain

#### Advantages
- Short and clean syntax
- Faster execution
- Improves readability
- Easy to filter and transform data
- Less code compared to loops

#### Limitations
- Suitable only for simple logic
- Complex expressions reduce readability
- Not suitable for multiple statements

#### Difference Between for Loop and List Comprehension

| For Loop                 | List Comprehension      |
| ------------------------ | ----------------------- |
| Multiple lines           | Single line             |
| Uses `append()`          | Creates list directly   |
| Better for complex logic | Better for simple logic |
| More code                | Less code               |
| Slightly slower          | Generally faster        |


#### ⭐ Interview Questions
- **What is List Comprehension?**
- A one-line method to create a list using a for loop.
 
- **Why is List Comprehension faster?**

- Because it is optimized internally in Python and reduces the overhead of repeatedly calling append().

**It is commonly used for:**

- Creating new lists
- Filtering data
- Transforming values
- String processing
- Working with nested lists

```python
# syntaxs
case-1:[output forloop <cond>]
case-2:[output forloop if<cond>]
case-3:[output1 if<cond1> else<output2> forloop]
case-4:[ouput1 if con1,else,output2 if con2,else,output3 if con3,else ,output4  for i in list]
```

#### 2.Syntax
**Basic Syntax**

```python
[expression for variable in iterable]
```

##### Components
- expression → Value to be stored in the list
- variable → Loop variable
- iterable → List, Tuple, Set, Dictionary, String, Range, etc.

#### 3. Basic Examples
- **Example 1: Create a List**
- **Without List Comprehension**

```python
num=[]
for i in range(1,6):
    num.append(i)
print(num)
```

**Using List Comprehension**

```python
nums=[i for i in range(1,6)]
nums
```

#### valrus operator(:=)
- Walrus Operator (:=) is used to assign a value to a variable as part of an expression, reducing the need for separate assignment statements.

```python
l=range(1,11)
s=0
l=[s:=s+i for i in range(1,11)]
l
```
