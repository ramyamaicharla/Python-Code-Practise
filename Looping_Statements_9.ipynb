### Loops in Python

- Loops are used to repeat a block of code multiple times without writing the same code again.

#### What is a Loop?
#### Definition

- A loop is a programming structure that repeatedly executes a block of code until a condition becomes false or until all items in a sequence have been processed.

#### Simple Definition

  - **A loop is used to execute the same block of code multiple times**.

#### Why Do We Use Loops?

```python
# Without loops:
print("Python")
print("Python")
print("Python")
print("Python")
print("Python")
```

```python
# using a loop
for i in range(5):
    print("python")

# The output is the same, but the code is shorter and easier to maintain.
```

#### Types of Loops in Python

#### 1.while Loop
#### 2.for Loop

### 1. while Loop
- The while loop executes a block of code **as long as the given condition is True**.

```python
# syntax
while condition:
    statements
```

```python
# Flow Diagram
Start
   │
   ▼
Condition
   │
True ─────► Execute Code
   ▲             │
   │             ▼
   └─────────────┘
False
   │
   ▼
 End
```

```python
#Example 1
#Print numbers from 1 to 5.
i=1
while i<=5:
    print(i)
    i=i+1
```

```python
# Infinite Loop
while True:
    print("Python")
# Warning: This loop never ends unless you stop it manually (Ctrl + C).    
```

### 2. for Loop

- The for loop is used to iterate over a sequence such as a string, list, tuple, set, dictionary, or range.

```python
# Syntax
for variable in sequence:
    statements
```

### Difference Between for and while

| `for` Loop                                  | `while` Loop                                  |
| ------------------------------------------- | --------------------------------------------- |
| Used when the number of iterations is known | Used when the number of iterations is unknown |
| Simpler for sequences                       | Better for condition-based repetition         |


#### Nested Loops

- A nested loop is a loop inside another loop.

```python
# Syntax
for i in range(...):
    for j in range(...):
        statements
```

```python
# example
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

#### Pattern Example

```python
for i in range(5):
    for j in range(5):
        print("*", end=" ")
    print()
```

#### Key Points to Remember
- Use a for loop when you know how many times you want to repeat.
- Use a while loop when repetition depends on a condition.
- break exits the loop immediately.
- continue skips the current iteration and continues with the next one.
- pass is a placeholder that performs no action.
- Nested loops are commonly used for patterns, tables, and matrix-like operations.

**Jumping Statements**

- **1.break**— The break statement immediately stops the loop.
- **2.continue** — The continue statement skips the current iteration and moves to the next iteration.
- **3.pass** — The pass statement is a placeholder. It does nothing and is used when a statement is syntactically required but no action is needed

```python
for i in range(1,11):
    if i==6:
        break
    print(i)
```

```python
for i in range(1,11):
    if i==5:
        continue
    print(i)
```

```python
for i in range(1,11):
    if i==7:
        continue
    print(i)
```

| Statement  | Purpose                     |
| ---------- | --------------------------- |
| `break`    | Stops the loop completely   |
| `continue` | Skips the current iteration |
| `pass`     | Does nothing (placeholder)  |
