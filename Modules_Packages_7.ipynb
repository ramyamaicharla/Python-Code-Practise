### Modules & Packages in Python

**import Statement**
- Definition

- The import statement is used to use a module in another Python program.

```python
# Syntax
import module_name
```

```python
# example
import math
math.sqrt(25)
```

### Import Multiple Modules

```python
import math
import random

------(or)------
import math,random
```

#### Import with Alias

- Sometimes module names are long.

- Use **as**.

```python
import math as m
print(m.sqrt(64))
```

#### from Statement
- Definition

- The from statement imports only the required function or variable from a module.

```python
# syntax
from module_name import function
```

```python
# example
from math import sqrt
sqrt(36)
```

### Import Multiple Functions

```python
from math import sqrt,factorial
```

#### Import Everything

```python
from import math*
# Now you can directly write
```

```python
print(sqrt(25))
# instead of
math.sqrt(25)

# Note: Avoid from module import * in large projects because it can make code harder to read and may cause name conflicts.
```

### Difference Between import and from

| `import`                  | `from`                                     |
| ------------------------- | ------------------------------------------ |
| Imports the entire module | Imports specific functions or variables    |
| Use `module.function()`   | Use the function directly                  |
| Better for large programs | Convenient when using only a few functions |


```python
# example
import math
math.sqrt(23)
```

```python
from math import sqrt
sqrt(23)
```

### Structure of a Package

```python
mypackage/
│
├── __init__.py
├── calculator.py
├── student.py
└── employee.py
```

#### Why __init__.py?

- __init__.py tells Python that the folder should be treated as a package.

- **Note**: In modern Python versions, packages can often work without __init__.py, but including it is still considered good practice, especially for beginners and structured projects

### What is pip?
- Definition

- pip stands for **Pip Installs Packages**.

- It is Python's package manager.

- Simple Definition

   - pip is a tool used to install, update, and remove Python packages.

### Why Do We Use pip?

- Python does not include every library.

- Use pip to install additional packages.

- Examples

  - NumPy
  - Pandas
  - Flask
  - Django
  - Requests

### Check pip Version

```python
pip --version
```

###  Install a Package

```python
pip install numpy
pip install pandas
```

#### Install Multiple Packages

```python
pip install numpy pandas matplotlib
```

#### Upgrade a Package

```python
pip install --upgrade pandas
```

#### Uninstall a Package

```python
pip unistall numpy
```

#### List Installed Packages

```python
pip list
```

#### Show Package Information

```python
pip show pandas
```

#### Freeze Installed Packages

```python
pip freeze
```

#### Install from requirements.txt

```python
pip install -r requirements.txt
```

#### Create requirements.txt

```python
pip freeze > requirements.txt
```

### Module vs Package

| Module                                 | Package                             |
| -------------------------------------- | ----------------------------------- |
| A single `.py` file                    | A folder containing related modules |
| Contains functions, variables, classes | Contains multiple modules           |
| Example: `math.py`                     | Example: `mypackage`                |


### import vs from

| `import`                 | `from`                 |
| ------------------------ | ---------------------- |
| Imports the whole module | Imports selected items |
| `math.sqrt()`            | `sqrt()`               |


### Module vs Library vs Package

| Module             | Package               | Library                            |
| ------------------ | --------------------- | ---------------------------------- |
| Single Python file | Collection of modules | Collection of packages and modules |
| Example: `math.py` | Example: `mypackage`  | Example: NumPy, Pandas             |


#### Summary

#### Key Points to Remember
- **A module** is a single Python (.py) file containing reusable code.
- **A package** is a folder that contains related modules.
- import imports an entire module.
- from ... import ... imports specific functions or variables.
- **pip** is Python's package manager used to install and manage third-party packages.
- Use **user-defined modules** to organize your own code and packages to group related modules together.
