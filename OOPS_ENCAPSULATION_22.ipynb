### Encapsulation

- Bundling data + methods together, and restricting direct access to internal state

- Encapsulation means an object controls its own data. The internal variables (fields) are kept private, and the outside world can only interact with the object through carefully defined public methods. This prevents the object's data from being accidentally broken by outside code.

- Analogy: A capsule (medicine pill) — the ingredients inside are sealed and protected. You swallow it; you don't mix the chemicals yourself.

- Encapsulation means **hiding data** and allowing access only through methods.

### Why it matters

- Data protection

   - No outside code can set balance = -9999 directly. Only your controlled methods can touch it.
 
- Flexibility

   - You can change the internal implementation without breaking code that uses the class.
 
- Validation

   - Setters can check values before accepting them — e.g., reject a negative age.
 
- Modularity

    - Each class is a self-contained unit — easy to test and debug independently.

```python
# Advanced Real-World Challenges 
# 16. ATM Machine 
# Features: 
# • Private PIN  
# • Private Balance  
# Methods: 
# deposit() 
# withdraw() 
# change_pin() 
# check_balance() 
# Validation: 
# • Correct PIN required  
# • Withdraw <= Balance 

class ATM:
    def __init__(self, balance, pin):
        self.__atm_balance = balance
        self.__atm_password = pin

    def deposit(self, pin, amount):
        if pin == self.__atm_password:
            if amount > 0:
                self.__atm_balance += amount
                return "Amount deposited successfully"
        return "Invalid PIN"

    def withdraw(self, pin, amount):
        if pin == self.__atm_password:
            if 0 < amount <= self.__atm_balance:
                self.__atm_balance -= amount
                return "Amount withdrawn successfully"
            else:
                return "Insufficient balance"
        return "Invalid PIN"

    def change_pin(self, old_pin, new_pin):
        if old_pin == self.__atm_password:
            self.__atm_password = new_pin
            return "PIN changed successfully"
        return "Invalid PIN"

    def check_balance(self, pin):
        if pin == self.__atm_password:
            print("Balance:", self.__atm_balance)
        else:
            print("Invalid PIN")


atm = ATM(20000, 1234)

print(atm.deposit(1234, 5000))
print(atm.withdraw(1234, 3000))
print(atm.change_pin(1234, 5678))
atm.check_balance(5678)
```

```python
# 17. Netflix Account 
# Private Attributes: 
# __email 
# __password 
# __subscription_plan 
# Validation: 
# • Email contains "@"  
# • Password length >= 8  
# Methods: 
# change_password() 
# upgrade_plan() 
# display_account()

class NetflixAccount:
    def __init__(self, email, password, subscription_plan):

        if "@" in email:
            self.__email = email
        else:
            self.__email = "Invalid Email"
            print("Email must contain '@'")

        if len(password) >= 8:
            self.__password = password
        else:
            self.__password = ""
            print("Password length must be at least 8 characters")

        self.__subscription_plan = subscription_plan

    def change_password(self, new_password):
        if len(new_password) >= 8:
            self.__password = new_password
            return "Password changed successfully"
        else:
            return "Password must be at least 8 characters"

    def upgrade_plan(self, new_plan):
        self.__subscription_plan = new_plan
        return "Plan upgraded successfully"

    def display_account(self):
        print("Email:", self.__email)
        print("Subscription Plan:", self.__subscription_plan)


# Create object
nf = NetflixAccount("ramya@gmail.com", "ramya123", "Basic")

print(nf.change_password("newpass123"))
print(nf.upgrade_plan("Premium"))

nf.display_account()
```

```python
# 19. School Management System 
# Private Attributes: 
# __student_name 
# __roll_number 
# __fees 
# Methods: 
# pay_fees() 
# remaining_fees() 
# display_student() 
# Validation: 
# • Fees cannot be negative

class Student:
    def __init__(self, student_name, roll_number, fees):
        self.__student_name = student_name
        self.__roll_number = roll_number

        if fees >= 0:
            self.__fees = fees
        else:
            self.__fees = 0
            print("Fees cannot be negative")

    def pay_fees(self, amount):
        if amount > 0:
            self.__fees -= amount
            return "Fees paid successfully"
        else:
            return "Invalid amount"

    def remaining_fees(self):
        return self.__fees

    def display_student(self):
        print("Student Name:", self.__student_name)
        print("Roll Number:", self.__roll_number)
        print("Remaining Fees:", self.__fees)


# Create object
std = Student("Ramya", 101, 50000)

print(std.pay_fees(10000))
print("Remaining Fees:", std.remaining_fees())

std.display_student()
```
