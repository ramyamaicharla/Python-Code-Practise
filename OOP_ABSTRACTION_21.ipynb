#### 8. Abstraction
**What is Abstraction?**
- hiding implementation details
- Showing only essential details and hiding implementation details.
- it focuses on **what an object does**,not how it does it
- using **abstact class**

Real-Time Example

- You drive a car without knowing how the engine works.

- Abstraction means **hiding implementation details** and showing only the required functionality.

```python
# both encapsulation and abstraction
from abc import ABC, abstractmethod

# parent abstract method
class ATM(ABC):
    def __init__(self,account_number,balance):
        self.__atm_account_number=account_number

        if balance>=0:
            self.__atm_balance=balance
        else:
            self.__atm_balance=0
        
    #accessor method   # read the private data
    def get_account_num(self):
        return self.__atm_account_number

    def get_balance(self):
        return self.__atm_balance

    #setter method   # updating
    def set_balance(self,balance):
        self.__atm_balance=balance
        
    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self,amount):
        pass

class Child(ATM):

    def deposit(self,amount):
        if amount>0:
            self.set_balance(self.get_balance()+amount)
            print("Deposited amount:",amount)
        else:
            print("Deposit amount must be greater than 0")


    def withdraw(self,amount):
        if amount>0:
            if amount<=self.get_balance():
                self.set_balance(self.get_balance()-amount)
                print("Withdrawn Amount:", amount)
            else:
                print("Insufficient Balance")
        else:
            print("Withdrawal amount must be greater than 0")


atm = Child(12345,10000)
print('account number:',atm.get_account_num())
print('Balance:',atm.get_balance())

atm.deposit(20000)
print('balance after deposite:',atm.get_balance())

atm.withdraw(10000)
print("Balance after withdrawal:", atm.get_balance())

atm.withdraw(20000)
```

```python
from abc import ABC, abstractmethod

# parent abstract class
class Student(ABC):
    def __init__(self,name,marks):
        self.__std_name=name
        self.__std_marks=marks


    def get_std_name(self):
        return self.__std_name


    def get_std_marks(self):
        return self.__std_marks

    @abstractmethod
    def calculate_grade():
        pass

class SchoolStudent(Student):
    def __init__(self,name,marks):
        super().__init__(name,marks)


    def calculate_grade(self):
        marks=self.get_std_marks()

        if marks>=90:
            print("A")
        elif marks>=75:
            print("B")
        else:
            print("fail")


class CollegeStudent(Student):
    def __init__(self,name,marks):
        super().__init__(name,marks)

    def calculate_grade(self):
        marks = self.get_std_marks()
        
        if marks >= 85:
            return "Distinction"
        elif marks >= 70:
            return "First Class"
        elif marks >= 50:
            return "Second Class"
        else:
            return "Fail"

scl=SchoolStudent("sumit",90)
clg=CollegeStudent("ramya",72)

print("School Student")
print("Name :",scl.get_std_name ())
print("Marks:", scl.get_std_marks())
print("Grade:", scl.calculate_grade())

print("\nCollegeStudent")
print("Name :", clg.get_std_name())
print("Marks:", clg.get_std_marks())
print("Grade:", clg.calculate_grade())
    
```

```python
from abc import ABC, abstractmethod

# Abstract Parent Class
class Patient(ABC):
    def __init__(self, patient_name, bill_amount):
        self.__patient_name = patient_name
        self.__bill_amount = bill_amount

    # Getter Methods
    def get_patient_name(self):
        return self.__patient_name

    def get_bill_amount(self):
        return self.__bill_amount

    @abstractmethod
    def final_bill(self):
        pass


# Child Class 1
class InPatient(Patient):
    def __init__(self, patient_name, bill_amount):
        super().__init__(patient_name, bill_amount)

    def final_bill(self):
        room_charge = 2000
        return self.get_bill_amount() + room_charge


# Child Class 2
class OutPatient(Patient):
    def __init__(self, patient_name, bill_amount):
        super().__init__(patient_name, bill_amount)

    def final_bill(self):
        consultation_fee = 500
        return self.get_bill_amount() + consultation_fee


# Object Creation
in_patient = InPatient("Ramya", 10000)
out_patient = OutPatient("Priya", 3000)

print("In Patient Details")
print("Name:", in_patient.get_patient_name())
print("Bill Amount:", in_patient.get_bill_amount())
print("Final Bill:", in_patient.final_bill())

print("\nOut Patient Details")
print("Name:", out_patient.get_patient_name())
print("Bill Amount:", out_patient.get_bill_amount())
print("Final Bill:", out_patient.final_bill())
```

```python
from abc import ABC, abstractmethod

# Abstract Parent Class
class Product(ABC):
    def __init__(self, product_name, price):
        self.__product_name = product_name
        self.__price = price

    # Getter Methods
    def get_product_name(self):
        return self.__product_name

    def get_price(self):
        return self.__price

    @abstractmethod
    def discount(self):
        pass


# Child Class 1
class Electronics(Product):
    def __init__(self, product_name, price):
        super().__init__(product_name, price)

    def discount(self):
        discount_amount = self.get_price() * 0.15  # 15% discount
        return self.get_price() - discount_amount


# Child Class 2
class Clothing(Product):
    def __init__(self, product_name, price):
        super().__init__(product_name, price)

    def discount(self):
        discount_amount = self.get_price() * 0.25  # 25% discount
        return self.get_price() - discount_amount


# Object Creation
electronic_item = Electronics("Laptop", 50000)
clothing_item = Clothing("Shirt", 2000)

print("Electronics Product")
print("Product Name:", electronic_item.get_product_name())
print("Original Price:", electronic_item.get_price())
print("Price After Discount:", electronic_item.discount())

print("\nClothing Product")
print("Product Name:", clothing_item.get_product_name())
print("Original Price:", clothing_item.get_price())
print("Price After Discount:", clothing_item.discount())
```

```python
from abc import ABC, abstractmethod

# Abstract Parent Class
class Account(ABC):
    def __init__(self,account_holder,balance):
        self.__Account_holder =account_holder
        self.__Balance =balance

    def get_account_hoder(self):
        return self.__Account_holder

    def get_balance(self):
        return self.__Balance

    @abstractmethod 
    def calculate_interest():
        pass

class SavingsAccount(Account):
    def __init__(self,account_holder,balance):
        super().__init__(account_holder,balance)

    def calculate_interest(self):
        interest=self.get_balance()*0.05
        return interest


class CurrentAccount(Account):
    def __init__(self,account_holder,balance):
        super().__init__(account_holder,balance)

    def calculate_interest(self):
        interest=self.get_balance()*0.02
        return interest



savings = SavingsAccount("Ramya", 100000)
current = CurrentAccount("Priya", 100000)

print("Savings Account Details")
print("Account Holder:", savings.get_account_hoder())
print("Balance:", savings.get_balance())
print("Interest:", savings.calculate_interest())

print("\nCurrent Account Details")
print("Account Holder:", current.get_account_hoder())
print("Balance:", current.get_balance())
print("Interest:", current.calculate_interest())        
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, vehicle_name, rent_per_day):
        self.__vehicle_name = vehicle_name
        self.__rent_per_day = rent_per_day

    def get_vehicle_name(self):
        return self.__vehicle_name

    def get_rent_per_day(self):
        return self.__rent_per_day

    @abstractmethod
    def calculate_rent(self, days):
        pass


class Car(Vehicle):
    def __init__(self, vehicle_name, rent_per_day):
        super().__init__(vehicle_name, rent_per_day)

    def calculate_rent(self, days):
        return self.get_rent_per_day() * days + 500


class Bike(Vehicle):
    def __init__(self, vehicle_name, rent_per_day):
        super().__init__(vehicle_name, rent_per_day)

    def calculate_rent(self, days):
        return self.get_rent_per_day() * days + 100


car = Car("Swift", 1500)
bike = Bike("Activa", 500)

print(car.calculate_rent(3))
print(bike.calculate_rent(3))
```

```python
from abc import ABC, abstractmethod

class Book(ABC):
    def __init__(self, book_name, copies):
        self.__book_name = book_name
        self.__copies = copies

    def get_book_name(self):
        return self.__book_name

    def get_copies(self):
        return self.__copies

    @abstractmethod
    def issue_book(self):
        pass


class ReferenceBook(Book):
    def __init__(self, book_name, copies):
        super().__init__(book_name, copies)

    def issue_book(self):
        print("Reference books cannot be issued.")


class NormalBook(Book):
    def __init__(self, book_name, copies):
        super().__init__(book_name, copies)

    def issue_book(self):
        if self.get_copies() > 0:
            print("Book Issued Successfully")
        else:
            print("No Copies Available")


ref = ReferenceBook("Dictionary", 2)
normal = NormalBook("Python", 5)

ref.issue_book()
normal.issue_book()
```

```python
from abc import ABC, abstractmethod

class Electricity(ABC):
    def __init__(self, customer_name, units):
        self.__customer_name = customer_name
        self.__units = units

    def get_customer_name(self):
        return self.__customer_name

    def get_units(self):
        return self.__units

    @abstractmethod
    def calculate_bill(self):
        pass


class Domestic(Electricity):
    def __init__(self, customer_name, units):
        super().__init__(customer_name, units)

    def calculate_bill(self):
        return self.get_units() * 5


class Commercial(Electricity):
    def __init__(self, customer_name, units):
        super().__init__(customer_name, units)

    def calculate_bill(self):
        return self.get_units() * 8


d = Domestic("Ramya", 100)
c = Commercial("Priya", 100)

print(d.calculate_bill())
print(c.calculate_bill())
```

```python
from abc import ABC, abstractmethod

class Room(ABC):
    def __init__(self, room_type, price_per_day):
        self.__room_type = room_type
        self.__price_per_day = price_per_day

    def get_room_type(self):
        return self.__room_type

    def get_price_per_day(self):
        return self.__price_per_day

    @abstractmethod
    def calculate_charge(self, days):
        pass


class ACRoom(Room):
    def __init__(self, room_type, price_per_day):
        super().__init__(room_type, price_per_day)

    def calculate_charge(self, days):
        return self.get_price_per_day() * days + 1000


class NonACRoom(Room):
    def __init__(self, room_type, price_per_day):
        super().__init__(room_type, price_per_day)

    def calculate_charge(self, days):
        return self.get_price_per_day() * days + 500


ac = ACRoom("Deluxe AC", 3000)
nonac = NonACRoom("Standard", 2000)

print(ac.calculate_charge(2))
print(nonac.calculate_charge(2))
```

### Difference Between Encapsulation and Abstraction

| Encapsulation          | Abstraction           |
| ---------------------- | --------------------- |
| Hides data             | Hides implementation  |
| Uses private variables | Uses abstract classes |
| Improves security      | Reduces complexity    |
