**1.Duck typing**
- Duck Typing is a concept in Python where the type of an object is not important; what matters is whether the object has the required method or behavior.


```python
# Q1 — Easy 
# Create 3 classes Dog, Cat, Bird 
# • Each has a method speak() 
# • Dog → Woof!, Cat → Meow!, Bird → Tweet! 
# • Create a function make_sound(animal) 
# • Call it with all 3 objects in a loop

class Dog:
    def Speak(self):
        return  "Dog:wooff!....."

class Cat:
    def Speak(self):
        return "Cat:meow!...."

class Bird:
    def Speak(self):
        return "Bird:tweet!..."

def make_sound(animal):
    print(animal.Speak())


# object
animals=[Dog(),Cat(),Bird()]

#call function in aloop
for animal in animals:
    make_sound(animal)
```

**2.Method Overriding**

- Method Overloading is defining multiple methods with the same name but different parameters in the same class.

- Note: Python does not support true method overloading directly. It is usually achieved using default arguments or *args.

```python
class Animal:
    #method1
    def speak(self):
        print("***********")

    #method2
    def eat(self):
        print("eating dal batti")

# child class
class Cat(Animal):

    # method
    def speak(self):
        print("meowwwwwww")

    #method 2
    def eat(self):
        print("kaju bhadham")


#great the object
cat=Cat()
cat.eat()
```

**3.Method Overloading**


- Method Overriding occurs when a child class provides its own implementation of a method that already exists in the parent class.

- The child class method replaces (overrides) the parent class method.

- we can use method overloading,using bydefault 

| Method Overloading                     | Method Overriding                    |
| -------------------------------------- | ------------------------------------ |
| Same method name, different parameters | Same method name and same parameters |
| Usually in the same class              | In parent and child classes          |
| Compile-time polymorphism              | Run-time polymorphism                |
| Python supports it indirectly          | Python fully supports it             |


```python
#without method over loading
class cal:
    def add(self,a,b,c=2,d=2):
        return a+b+c+d


ob=cal()
ob.add(2,3)
ob.add(2,3,4)
ob.add(2,3,4,5)
```

**Operator Overloading**

- Operator Overloading is a feature of polymorphism that allows operators (+, -, *, ==, etc.) to perform different operations depending on the objects they are used with.

- In Python, operator overloading is achieved using special (magic/dunder) methods.

- dunder method

- ( __method__ )

| Operator | Magic Method    |
| -------- | --------------- |
| `+`      | `__add__()`     |
| `-`      | `__sub__()`     |
| `*`      | `__mul__()`     |
| `/`      | `__truediv__()` |
| `==`     | `__eq__()`      |
| `<`      | `__lt__()`      |
| `>`      | `__gt__()`      |


```python
print(10 + 20)        # 30
print("Ram" + "ya")   # Ramya
```

```python
class Student:
    #con
    def __init__(self,a,b):
        self.A=a
        self.B=b


    def add(self):
        return self.A+self.B


    # dunder method names
    def __add__(self,other):
        return(self.A+other.B)
        
     #method:sub
    def __sub__(self,other):
        return (self.A-other.B)

     #method:mul
    def __mul__(self,other):
        return (self.A*other.B)

     #method:div
    def __truediv__(self,other):
        return (self.A/other.B)

    #method:eq
    def __eq__(self,other):
        return (self.A==other.B) 

     #method:eq
    def __it__(self,other):
        return (self.A<other.B) 

     #method:eq
    def __gt__(self,other):
        return (self.A>other.B) 


ob1=Student(1,2)
ob2=Student(3,4)
ob3=Student(1,1)
print(ob1+ob3)
print(ob1-ob3)
print(ob1*ob3)
print(ob1/ob3)
print(ob1==ob3)
print(ob1<ob3)
print(ob1>ob3)
```

- The same + operator performs:

- Addition for numbers
- Concatenation for strings

- This is Operator Overloading.
