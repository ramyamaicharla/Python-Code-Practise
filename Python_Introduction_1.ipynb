### Translator Programs

- Computers understand only machine language.

- A translator converts human-readable code into machine language.

- There are three types:

    - Assembler
    - Compiler
    - Interpreter

**1.Assembler**
- An assembler converts Assembly Language (2GL) into machine language.

- Simple Definition

 - Assembler converts Assembly Language into machine language.

**2.Compiler**
- A compiler converts the **entire program** into **machine language** before execution.

                                        Entire Program
                                        
                                        ↓
                                        
                                        Compiler
                                        
                                        ↓
                                        
                                        Machine Code
                                        
                                        ↓
                                        
                                        Run

- Advantages
    - Fast execution
    - Finds many errors at once
- Disadvantages
    - Must compile the whole program before running it

**3.Interpreter**
- An interpreter converts and executes **one line of code at a time**.
- Python uses an interpreter.

                    Line 1
                    
                    ↓
                    
                    Execute
                    
                    ↓
                    
                    Line 2
                    
                    ↓
                    
                    Execute
                    
                    ↓
                    
                    Output

- Advantages
  - Easy to debug
  - Errors are shown immediately
- Disadvantages
  - Slower than a compiler

| Compiler                           | Interpreter                 |
| ---------------------------------- | --------------------------- |
| Converts the whole program at once | Converts one line at a time |
| Faster execution                   | Slower execution            |
| Errors shown after compilation     | Errors shown immediately    |
| Example: C, C++                    | Example: Python             |

## Introduction to Python

### 1.What is Python?
- Python is a **high-level**, **interpreted**,**object oriented**, **dynamically-typed programming language**.
- Simple, English-like syntax (easy to learn)
- Widely used in **Data Science, AI/ML, Web Development, Automation**
- Huge community support and libraries (pandas, NumPy, scikit-learn, etc.)
- why we using python only in data science?- beacuse **python has a large standard library**

### 2.Why is Python Popular?
- Easy to learn
- Easy to read

- Cross-platform

- Used in AI, Data Science, Automation, Web Development, and Machine Learni

### 3.History of Python
- Creator:
- Python was created **by Guido van Rossum, a Dutch programmer, in 1989 and released publicly in 1991**.
### Why it was created:
- Guido wanted to create a language that was:

- Easy to read and write (like English)
- Fun to code in **(named after "Monty Python's Flying Circus" comedy show, not the snake!)**
- A successor to a **language called ABC**, but more powerful and flexible

- Evolution:

- 1991 – Python 1.0 released (simple, basic features)
- 2000 – Python 2.0 released (added list comprehensions, garbage collection)
- 2008 – Python 3.0 released (major redesign, not fully compatible with Python 2)
- Today – Python 3.x is actively updated and is the industry standard

### Why Python?

- **Simplicity** : Code looks almost like plain English, so **it's easy to understand even for beginners**
- **Easy Syntax** : No complicated symbols like **{} or ; — uses indentation (spacing) to organize code**
- **Large Community** : **Millions of developers use Python worldwide, so help, tutorials, and solutions are easily available online (like Stack Overflow)**
- large standard library

### 4. Features of Python
- **Easy**: Simple to learn, even for non-programmers
- **Readable**: Code looks clean and understandable, like English sentences
- **Portable**: Write code once, run it on any OS (Windows, Mac, Linux) without changes
- **Free**: No cost to download or use
- **Open Source**: Source code is publicly available; anyone can view, modify, and contribute
- **Interpreted**: Code runs line-by-line (no separate compiling step needed)
- **Dynamic Typing**: No need to declare variable types; Python figures it out automatically
- **Object-Oriented**: Supports OOP concepts (classes, objects, inheritance, etc.)
- **Huge Libraries**: Has ready-made code packages for almost everything (data science, AI, web, etc.)
- **Cross Platform**: Works the same way on different operating systems
- **Extensible**: Can integrate with other languages like C, C++ for performance-heavy tasks
- **Embeddable**: Python code can be embedded inside other applications (like a script running inside software)
