## Sets in Python

### 1. Introduction
#### What is a Set?

- A **Set is an unordered collection of unique elements** in Python.

   - Sets do **not allow duplicate values**.
   - Sets are **mutable**, meaning you can add or remove elements.
   - Sets are enclosed in **curly braces {}**.

```python
# syntax
set_name = {value1, value2, value3}
```

```python
# example
numbers={10,20,30,40,50}
print(numbers)
# Note: The order may vary because sets are unordered.
```

#### Why Set?

- Sets are useful when:

  -  You want to store **unique values**.
  -  You need **fast membership testing**.
  -  You want to perform **mathematical set operations** like union and intersection.
  -  You want to remove duplicate values from a collection.

### Features of Sets
- Unordered
- Mutable
- Do not allow duplicate values
- Cannot access elements using indexing
- Supports mathematical set operations
- Can store different data types (immutable objects only)

```python
# example
data={10,20.5,True,"ramya"}
data
```

#### 3. Set Operations

```python
# assume
a={1,2,3,4}
b={3,4,5,6}
```

#### Union
- Combines all unique elements.

```python
# print(a|b)
print(a.union(b))
```

#### Intersection
- Returns common elements.

```python
#print(a&b)               # these both are same...u can use anyone method
print(a.intersection(b))
```

#### Difference
- Elements present in the first set but not in the second.

```python
# print(a-b)
print(a.difference(b))
```

#### Symmetric Difference
- Elements present in either set but not both.

```python
#print(a^b)
print(a.symmetric_difference(b))
```

### 4. Set Methods

#### add()
- Adds one element.

```python
nums={10,20}
nums.add(30)
nums
```

#### update()
- Adds multiple elements.

```python
numbers={10,20}
numbers.update([30,40,50])
print(numbers)
```

#### remove()
- Removes an element.

```python
s={1, 2, 5, 6}
s.remove(5)
s

# Note: Raises KeyError if the element is not found.
```

#### discard()
- Removes an element if it exists.

```python
s={1, 2, 5, 6}
s.discard(6)
print(numbers)

# No error if the element is absent.
```

### pop()
- Removes and returns a random element.

```python
numbers = {10,20,30}
remove=numbers.pop()

print("Removed:",remove )
print(numbers)

# Note: Since sets are unordered, the removed element is unpredictable.
```

#### clear()
- Removes all elements

```python
numbers = {10,20,30}
numbers.clear()
print(numbers)
```

#### copy()
- Creates a copy of a set.

```python
a={10,20,30}
b=a.copy()
print(b)
```

### 5. Membership Operators

**in**

```python
numbers={10,20,30}
print(20 in numbers)
```

**not in**

```python
numbers={10,20,30}
print(20 not in numbers)
print(40 not in numbers)
```

### 6. Frozen Set

- A **frozenset is an immutable version of a set**.

 - Cannot add elements.
 - Cannot remove elements.
 - Supports union, intersection, difference, etc.

#### Creating a Frozen Set

```python
fs=frozenset([10,20,30,40])
print(fs)
```

#### Attempt to Modify

```python
fs=frozenset([10,20,30,40])
fs.add(40)

# AttributeError:
# 'frozenset' object has no attribute 'add'
```

### Difference Between Set and Frozen Set
| Set                     | Frozen Set         |
| ----------------------- | ------------------ |
| Mutable                 | Immutable          |
| Can add/remove elements | Cannot modify      |
| Uses `set()`            | Uses `frozenset()` |


### Difference Between list and Frozen Set
| List              | Set           |
| ----------------- | ------------- |
| Allows duplicates | No duplicates |
| Ordered           | Unordered     |
| Indexing possible | No indexing   |


### 7. Common Set Programs

#### Remove Duplicates

```python
nums=[10,20,30,40,50,10,20,10,10]
list(set(nums))
```

### Find Common Elementsm

```python
A = {10,20,30,40}
B = {30,40,50}
print(A.union(B))
```

### Find Unique Elements

```python
A = {10,20,30,40}
B = {30,40,50}
print(A^B)
```

### Set Difference

```python
A = {10,20,30,40}
B = {30,40,50}
print(A-B)
```

### Set Union

```python
A = {10,20,30,40}
B = {30,40,50}
print(A|B)
```

### Check Whether Two Sets are Equal

```python
A = {10,20,30}
B = {30,40,50}

if A==B:
    print("sets are equal")
else:
    print("sets are not equal")
```

### Find Elements Present Only in the Second Set

```python
A = {10,20,30}
B = {20,30,40,50}
print(B - A)
```

### ⭐ Summary
- A **set is an unordered collection of unique elements**.
- Sets **do not allow duplicates and do not support indexing or slicing**.
- Create an empty set using set(), not {}.
- Common operations include **union (|), intersection (&), difference (-), and symmetric difference (^)**.
- Frequently used methods are add(), update(), remove(), discard(), pop(), clear(), and copy().
- Use **in and not in** to check whether an element exists in a set.
- A **frozenset is an immutable set** that cannot be modified after creation.
- Sets are commonly used to **remove duplicates**, **find common elements**, **identify unique elements**, and perform efficient mathematical set operations.
