### Introduction to Strings

#### 1. What is a String?
- A String is a collection (sequence) of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

- A character can be:

 - Letters (A-Z, a-z)
 - Numbers (0-9)
 - Symbols (@, #, $, %, etc.)
 - Spaces

- Simple Definition

  - A string is a sequence of characters enclosed in quotes.

### 3. Negative Indexing
- Negative indexing starts from -1 and moves from right to left.

```python
String :    P    y    t    h    o    n    #right to left
Positive :  0    1    2    3    4    5
Negative : -6   -5   -4   -3   -2   -1
```

#### Positive vs Negative Indexing

| Positive      | Negative       |
| ------------- | -------------- |
| Starts from 0 | Starts from -1 |
| Left → Right  | Right → Left   |


#### Key Points to Remember
- Indexing accesses a single character.
- Positive indexing starts at 0 and moves left to right.
- Negative indexing starts at -1 and moves right to left.
- Slicing extracts a part of a string using start:stop.
- The stop index is excluded in slicing.
- The step value controls how many positions to move.
- [::-1] is the simplest and most common way to reverse a string in Python.
- Trying to access an invalid index raises an IndexError.

#### 6. Iteration Using for
- Definition

 - A for loop is used to access each character of a string one by one

```python
# syntax
for variable in string:
    statements
```

```python
name="manikanta"
for i in name:
    print(i)
```

```python
# iteration with index
name="manikanta"
for i in range(len(name)):
    print(i,name[i])
```

```python
# while loop
name = "Python"

i = 0
while i < len(name):
    print(name[i])
    i += 1
```

```python
# while reverse
name = "Python"

i = len(name) - 1
while i >= 0:
    print(name[i])
    i -= 1
```

#### 2. Searching Methods

- These methods search for characters or substrings.

| Method       | Purpose                               |
| ------------ | ------------------------------------- |
| find()       | First occurrence                      |
| rfind()      | Last occurrence                       |
| index()      | First occurrence (error if not found) |
| rindex()     | Last occurrence (error if not found)  |
| count()      | Counts occurrences                    |
| startswith() | Checks beginning                      |
| endswith()   | Checks ending                         |


```python
#find()
# Returns the index of the first occurrence.
text="python programming"
print(text.find("o"))
```

```python
#rfind()
# Returns the index of the last occurrence.
text="programming"
print(text.rfind("m"))
```

```python
# index()
# Similar to find(), but raises an error if not found.
text="python"
print(text.index("t"))
```

```python
# rindex()
# Returns the last occurrence
text="programming"
print(text.rindex("m"))
```

```python
# count():
# Counts how many times a substring appears.
text="banana"
print(text.count("a"))
```

```python
# startswith()
text="ramya"
print(text.startswith("r"))
```

```python
# endswith()
text="ramya"
print(text.startswith("ya"))
```
