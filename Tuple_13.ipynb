### Tuples in Python

### 1. Introduction
#### What is a Tuple?

- A Tuple is an ordered collection of elements in Python.

   - Tuples can store multiple values in a single variable.
   - Tuples are immutable, meaning their elements cannot be changed after creation.
   - Tuples can store different data types.

```python
# syntax
tuple_name = (value1, value2, value3)
```

```python
# example
numbers=(1,2,3,4,5,6)
numbers
```

```python
type(numbers)
```

#### Why Tuple?

- Tuples are useful when:

   - Data should not be modified.
   - Faster than lists.
   - Used as dictionary keys (lists cannot be).
   - Protects data from accidental changes.

#### Features of Tuples
- Ordered
- Immutable
- Allows duplicate values
- Supports indexing and slicing
- Can store multiple data types
- Faster than lists

```python
# example
t1=(10,20,3.96,True,"ramya",46,10,20,"apple")
t1
```

#### 3. Tuple Operations

```python
colors=("red","green","black","blue")
print(colors[0])
print(colors[1])
print(colors[2])
```

**Negative Indexing**

```python
colors=("red","green","black","blue")
print(colors[-1])
print(colors[-2])
print(colors[-3])
```

**Slicing**

```python
# syntax
tuple(start:stop:step)
```

```python
numbers = (10,20,30,40,50)

print(numbers[1:4])
print(numbers[:3])
print(numbers[2:])
print(numbers[::-1])
```

#### Iteration

#### 4. Tuple Methods
- Unlike lists, tuples have only two methods.

**count()**
- Returns the number of occurrences of an element

```python
nums=(1,2,3,4,5,6,8,9,10,2,2,2)
nums.count(2)
```

**index()**
- Returns the index of the first occurrence

```python
nums=(1,2,3,4,5,6,8,9,10)
print(nums.index(6))
```

#### 6. Packing & Unpacking
**Tuple Packing**
- Storing multiple values into one tuple.

```python
std=("ramya",22,"hyderabad")
print(std)
```

#### Tuple Unpacking
- Assign tuple values to variables.

```python
std=("ramya",22,"hyderabad")
name,age,city=std
print(f"student name:{name}")
print(f"age         :{age}")
print(f"city        :{city}")
```

#### Multiple Assignment

```python
a,b,c=10,20,30
print(a)
print(b)
print(c)
```

#### Swapping Variables

```python
print("\nBefore Swapping:")
a=int(input("enter a:"))
b=int(input("enter b:"))
a,b=b,a
print("\nAfter Swapping:")
print(f"a:{a}")
print(f"b:{b}")
```

#### 7. Immutable Nature
- A tuple cannot be modified after creation.

```python
numbers=(10,20,30,40,50)
numbers[1]=100

# TypeError:
# 'tuple' object does not support item assignment
```

### Why Tuples Cannot Be Modified?
- Because tuples are immutable.

- Advantages:

   - Faster than lists
   - Safer data storage
   - Can be used as dictionary keys
   - Less memory usage

#### Difference Between List and Tuple
| List                     | Tuple                                             |
| ------------------------ | ------------------------------------------------- |
| Mutable                  | Immutable                                         |
| Uses []                  | Uses ()                                           |
| Many methods             | Only 2 methods                                    |
| Slower                   | Faster                                            |
| Cannot be dictionary key | Can be dictionary key (if elements are immutable) |


### Difference Between Set and Tuple


| Set           | Tuple              |
| ------------- | ------------------ |
| Unordered     | Ordered            |
| No duplicates | Duplicates allowed |
| Mutable       | Immutable          |
| No indexing   | Indexing possible  |

#### ⭐ Summary
- A **tuple** is an **ordered, immutable collection** of elements.
- Tuples are created using **parentheses ()**.
- A **single-element tuple** must include a comma: (10,).
- Tuples support **indexing, negative indexing, slicing,
  iteration,concatenation, repetition, and membership testing**.
- Tuples provide only **two built-in methods**: count() and index().
- Common built-in functions such as len(), max(), min(), sum(), and
  sorted() work with tuples.
- **Tuple packing** stores multiple values in one tuple, **while tuple
  unpacking** assigns tuple values to multiple variables.
- Tuples are **immutable**, making them faster, memory-efficient, and suitable for storing fixed data.
- Common tuple programs include finding the **maximum, minimum, sum, searching elements, and converting between lists and tuples**.

```python
# process
# t = (
#       (10,20,30),      ← t[0]  # rows
#       (40,50,60),      ← t[1]
#       (70,80,90)       ← t[2]
#     )

#          0   1   2
# t[0] → (10, 20, 30)   # columns
#          ↑
#        t[0][0] = 10
#        t[0][1] = 20
#        t[0][2] = 30
```
