### Type Casting in Python

- In Python, type casting (or type conversion) means converting a value from one data type to another data type.

#### Common Type Casting Functions

| Function  | Converts To | Example                           |
| --------- | ----------- | --------------------------------- |
| `int()`   | Integer     | `int("10")` → `10`                |
| `float()` | Float       | `float("10.5")` → `10.5`          |
| `str()`   | String      | `str(100)` → `"100"`              |
| `bool()`  | Boolean     | `bool(1)` → `True`                |
| `list()`  | List        | `list("abc")` → `['a', 'b', 'c']` |
| `tuple()` | Tuple       | `tuple([1, 2, 3])` → `(1, 2, 3)`  |
| `set()`   | Set         | `set([1, 1, 2])` → `{1, 2}`       |


**1.Explicit casting**- is when you use functions like int(), float(), and str().


**2.Implicit casting**- is when Python automatically converts compatible types during operations.

#### 1.String to Integer

```python
x="25"
y=int(x)

print(y)   # 25
print(type(y)) # <class int>
```

#### 2.Integer to Float

```python
x=10
y=float(x)

print(y)
```

#### 3.Number to String

```python
x=21
text=str(x)

print("age:"+text)
```

#### 4.Integer to Boolean

```python
print(bool(0)) # False
print(bool(5)) # True
```

#### 5.Implicit Type Casting
- Python sometimes converts types automatically:

```python
x=5
y=2.5

result=x+y
print(result)
print(type(result))
```

#### 6.Handling Invalid Conversions

```python
x="hello"
# int(x)  # Raises ValueError
# Use try-except to avoid errors:
```

```python
try:
    num=int("hello")
except ValueError:
    print("cannot convert to integer")
```
