```python
# 10. Write a program that asks the user to enter a word and 
#then capitalizes every other letter of that word.
# So if the user enters rhinoceros,
# the program should print rHiNoCeRoS.

word = input("Enter a word: ")
new = ""

for i in range(len(word)):
    if i % 2 == 1:                        
        new = new + word[i].upper()       
    else:
        new = new + word[i]
print(new)
```

```python
# 14. Finding a substring within a string
# For example, if we were presented a series of lines formatted as follows:
# From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008
# and we wanted to pull out only the second half of the address (i.e., uct.ac.za)

s = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
domain = ""
start = False

for i in s:
    if i == "@":
        start = True
        continue
        
    if start:
        if i == " ":
            break
            
        domain = domain + i
print(domain)
```
