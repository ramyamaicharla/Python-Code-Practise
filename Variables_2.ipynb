### Variables

### 1. What is a Variable?

- A variable is a name given to a memory location that stores data.

- Simple Definition

  - A variable is a container used to store data in a program.

- Real-Life Example

 - Think of a water bottle.

    - Bottle =Variable
    - Water = Data

```python
age=20
name="ramya"
```

- Here:

- age is a variable.
- 20 is the value.
- name is a variable.
- "Ramya" is the value.

### 2. Why Variables?

- Variables are used to store information so that we can use it later in the program.

- Without variables, we would have to write the same value again and again.

```python
# without variables
print("Ramya")
print("Ramya")
print("Ramya")
```

```python
# with variables
name="ramya"
print(name)
print(name)
print(name)
```

#### Advantages of Variables
- Store data
- Reuse data
- Make programs easier to read
- Reduce repeated code
- Easy to update values

### 3. Memory Allocation

- When you create a variable, Python stores its value in the computer's memory (RAM).

```python
age=20
```

```python
Variable      Memory
 age  ------> 20

# You don't have to manage memory yourself—Python does it automatically.
```

### 4. Variable Naming Rules

- A variable name must follow these rules:

#### Rule 1: Start with a letter or underscore

```python
# correct
name="ramya"
_age=20
```

```python
# incorrect
2name = "Ramya"
```

#### Rule 2: Cannot start with a number

```python
123name = "Ramya"
```

#### Rule 3: No spaces

```python
# incorrect
student name = "Ramya"
```

```python
# correct
student_name="ramya"
```

#### Rule 4: Special characters are not allowed

```python
name@ = "Ramya"
```

#### Rule 5: Python keywords cannot be used

```python
if = 10
```

#### Rule 6: Variable names are case-sensitive

```python
# both are correct
# These are two different variables
age = 20
Age = 30
```

### 5. Identifiers

- An identifier is the name used to identify variables, functions, classes, and other objects.

- Simple Definition

   - An identifier is the name given to a variable, function, class, or object.

```python
student_name = "Ramya"
# student_name is an identifier.
```

### 6. Assignment

- The assignment operator (=) assigns a value to a variable

```python
age=20
# Store the value 20 in the variable age
```

### 7. Multiple Assignments

- Python allows assigning multiple variables in one line.

```python
a, b, c = 10, 20, 30

print(a)
print(b)
print(c)
```

```python
# Assign the same value to multiple variables.
# All variables store 100
x=y=z=100
```

### 8. Dynamic Typing

- Python automatically determines the data type.

- You don't need to declare it.

```python
age = 20
# Python knows it's an integer.
```

```python
# You can even change the type:
# This is allowed in Python
x = 10
x = "Hello"
```

### 9. Variable Swapping

- Swapping means exchanging values between variables.

#### Traditional Method

```python
a=10
b=20

temp=a
a=b
b=temp
```

#### Python Method

```python
a = 10
b = 20
a, b = b, a

print(a)
print(b)
```

### 10. Memory Address

- Every value stored in memory has a unique memory address.

```python
x = 10

# The value 10 is stored at some memory location.
# Python lets you view this address using the id() function
```

### 11. id() Function
- The id() function returns the memory address (identity) of an object.

```python
x=10
print(id(x))
# The exact number will be different on your computer
```

### 12. type() Function

```python
age=20
type(age)
```

```python
age=20
print(type(age))
```

```python
name = "Ramya"
print(type(name))
```

### 13. del Keyword
- The del keyword deletes a variable.

```python
x = 100
del x
```

### 15. Output Formatting
- Output formatting makes output look neat and readable.

#### Method 1: Comma

```python
name="ramya"
print("name:",name)
```

#### Method 2: format()

```python
name="ramya"
print("my name is {}".format(name))
```

#### Method 3: f-string (Recommended)

```python
name="Ramya"
age=21
print(f"my name is {name} and i am {age} years old.")
```
