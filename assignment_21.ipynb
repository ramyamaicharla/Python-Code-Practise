```python
# 1. Handle ValueError 
# Problem 
# Ask the user to enter a number. If the user enters text instead of a number, display: 
# Please enter a valid number! 
# Concepts 
# • try  
# • except  
# • ValueError 

try:
    num=int(input("enter a value:"))
    print(f"you entered :{num}")
except ValueError:
    print("plase enter the valid number!")
```

```python
# 2. Handle ZeroDivisionError 
# Problem 
# Take two numbers from the user and divide them. 
# If the denominator is 0, print: 
# Cannot divide by zero! 
# Concepts 
# • ZeroDivisionError  
# • Exception Handling 

try:
    n1=int(input("enter a n1:"))
    n2=int(input("enter a n2:"))

    result=n1/n2
except ZeroDivisionError:
    print("Cannot divide by zero!")
    
```

```python
# 3. Multiple Exceptions 
# Problem 
# Take a number from the user and divide 100 by it. 
# Handle: 
# • ValueError  
# • ZeroDivisionError  
# Concepts 
# • Multiple except blocks

try:
    num=int(input("enter a number:"))
    result=100/num
except ValueError:
    print("value error!")
except ZeroDivisionError:
    print("ZeroDivisionError!")
```

```python
# 5. File Not Found 
# Problem 
# Open a file named: 
# student.txt 
# If the file does not exist, display: 
# File not found! 
# Concepts 
# • FileNotFoundError 

try:
    file=open("student.txt","r")
    result=file.read()
    print(result)
except FileNotFoundError:
    print(f"error:file not found error")
```

```python
#6. Use else Block
try:
    num=int(input("enter a number:"))

except ValueError:
    print("Please enter a valid number.")

else:
    square = num ** 2
    print("Square of number is:", square)
```

```python
#7. Use finally Block 
try:
    file=open("std.txt","r")
    data=file.read()
    print(data)

except FileNotFoundError:
    print("file not found error!")
finally:
    print("read the content")
```

```python
# UnboundLocalError Handling Example

x = 10

def show():
    try:
        print(x)   
        x = 20     
    except UnboundLocalError:
        print("Error: Local variable referenced before assignment.")

show()
```

```python
# Custom Exception
class InsufficientBalanceError(Exception):
    pass


balance = 5000

try:
    amount = float(input("Enter withdrawal amount: "))

    if amount > balance:
        raise InsufficientBalanceError(
            "Insufficient balance in your account."
        )

except ValueError:
    print("Invalid input! Please enter a number.")

except InsufficientBalanceError as e:
    print(e)

else:
    balance -= amount
    print("Withdrawal Successful.")
    print("Remaining Balance:", balance)

finally:
    print("Transaction Completed.")
```
