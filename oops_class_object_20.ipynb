```python
class Calculator:
    def __init__(self,a,b):
        self.A=a
        self.B=b
    def add(self):
        print("Add:",self.A+self.B)

    def sub(self):
        print("Add:",self.A-self.B)

    def mul(self):
        print("Add:",self.A*self.B)

    def div(self):
        print("Add:",self.A/self.B)

cal=Calculator(10,20)
cal.add()
cal.sub()
cal.mul()
cal.div()
```

```python
# 1. Student Name Update 
# Question: 
# Create a Student class with attributes name and age. 
# Create an object and update the student's name from "Rahul" to "Amit". 
# What you need to do: 
# • Create a class.  
# • Create an object.  
# • Change the value of an attribute.

class Student:
    def __init__(self,name,age):
        self.Name=name
        self.Age=age

    def student_details(self):
        print("student name:",self.Name)
        print("student age :",self.Age)

    # updated
    def update_name(self,new_name):
        self.Name=new_name
        print("student new name:",self.Name)

# before update
std1=Student("rahul",20)
std1.student_details()

# after updating
std1.update_name("Amit")
```

```python
# 2.Car Speed Check 
# Question: 
# Create a Car class with attributes brand and speed. 
# If speed is greater than 80, print: 
# "Over Speed" 
# Otherwise print: 
# "Safe Driving" 
# What you need to learn: 
# • Access object attributes.  
# • Use if condition.

class Car:
    def __init__(self,brand,speed):
        self.Brand=brand
        self.Speed=speed
        
    def display_deatils(self):
        print("car brand:",self.Brand)
        
        if self.Speed>80:
            print("Over Speed")
        else:
            print("Safe Driving")

car1=Car('Tar',60)
car1.display_deatils()
```

```python
# 3. Employee Salary Update 
# Question: 
# Create an Employee class with attributes name and salary. 
# Increase salary by 5000 and display the updated salary. 
# Example: 
# Salary = 25000 
# Updated Salary = 30000 

class Employee:
    def __init__(self,name,salary):
        self.Name=name
        self.Salary=salary

    def display_details(self):
        print("employee name:",self.Name)
        print("employee salary:",self.Salary)
        
    def updated_salary(self):
        self.Salary+=5000
        print("updated salary:",self.Salary)

emp=Employee("Ramya",25000)
emp.display_details()
emp.updated_salary()
```

```python
# 4. Product Stock Check 
# Question: 
# Create a Product class with attributes: 
# product_name 
# stock 
# If stock is 0: 
# Out of Stock 
# Otherwise: 
# Available

class Product:
    def __init__(self,p_name,stock):
        self.P_name=p_name
        self.Stock=stock

    def display_details(self):
        print("product name:",self.P_name)
        print("product stock:",self.Stock)
        
        if self.Stock==0:
            print("Out of Stock")
        else:
            print("Available")



prd=Product("watch",10)
prd.display_details()
    
```

```python
# 5. Mobile Battery Status 
# Question: 
# Create a Mobile class with attributes: 
# model 
# battery 
# If battery is below 20: 
# Low Battery 
# Else: 
# Battery OK

class Mobile:
    def __init__(self,model,battery):
        self.mbl_model=model
        self.mbl_battery=battery

    def display(self):
        print("mobile model:",self.mbl_model)
        print("mobile battery:",self.mbl_battery)

        if self.mbl_battery<20:
            print("Low Battery")
        else:
            print("Battery OK")

mbl=Mobile("vivo",30)
mbl.display()
```

```python
# 7. Student Result System 
# Question: 
# Create a Student class with: 
# name 
# marks 
# Condition: 
# marks >= 90 → Grade A 
# marks >= 75 → Grade B 
# marks >= 50 → Grade C 
# otherwise Fail 
# Concepts: 
# • Multiple conditions (if-elif-else) 

class Student:
    def __init__(self,name,marks):
        self.std_name=name
        self.std_marks=marks

    def Student_Result_System(self):
        print('student name:',self.std_name)

        if self.std_marks>=90:
            print(f'Grade: {"A"}')
        elif self.std_marks>=75:
            print(f'Grade: {"B"}')
        elif self.std_marks>=50:
            print(f'Grade: {"C"}')
        else:
            print('fail')

std1=Student('sumit',90)
std1.Student_Result_System()
```

```python
# 8. Laptop Price Discount 
# Question: 
# Create a Laptop class with: 
# brand 
# price 
# If price is greater than 50000: 
# Apply 10% discount. 
# Display updated price. 

class Laptop:
    def __init__(self,brand,price):
        self.lap_brand=brand
        self.lap_price=price

    def  Laptop_Price_Discount(self):
        print('lap brand:',self.lap_brand)
        
    def discount(self):
         if self.lap_price>50000:
            self.lap_price -= self.lap_price * 10 / 100
            print('updated price:', self.lap_price)

lap1 = Laptop('Dell', 60000)
lap1.Laptop_Price_Discount()
lap1.discount()        
```

```python
# 9. Movie Ticket Booking 
# Question: 
# Create a MovieTicket class with: 
# name 
# age 
# Condition: 
# Age >= 18 → Allowed 
# Age < 18 → Not Allowed 
# After checking, update age by adding 1 year and display it.

class  MovieTicket:
    def __init__(self,name,age):
        self.per_name=name
        self.per_age=age

    def  Movie_Ticket_Booking(self):
        print('name of the person:',self.per_name)

        if self.per_age>=18:
            print('allowed')
        else:
            print('not allowed')

    def updating_age(self,new_age):
        self.per_age+=new_age
        print('updated age is:',self.per_age)

mov1=MovieTicket('sumit',29)
mov1.Movie_Ticket_Booking()   
```

```python
# 10. Online Shopping Cart 
# Question: 
# Create a ShoppingCart class with: 
# customer_name 
# total_amount 
# If total amount is greater than ₹1000: 
# Apply 20% discount 
# Otherwise: 
# Apply 5% discount 
# Display: 
# • Original Amount  
# • Discount  
# • Final Amount 

class  ShoppingCart:
    def __init__(self,cust_name,cust_amount):
        self.customer_name=cust_name
        self.total_amount=cust_amount

    def Online_Shopping_Cart(self):
        print('customer name:',self.customer_name)
        print('oroginal amount:',self.total_amount)
        
        if self.total_amount>1000:
            discount= self.total_amount * 20 / 100
        else:
            discount= self.total_amount * 5 / 100
        final_amount = self.total_amount - discount
        print('Discount:',discount)
        print('final price:',final_amount)

shp1=ShoppingCart('sumit sir',70000)
shp1.Online_Shopping_Cart()
```

**OOPS (Object-Oriented Programming System) – Definitions Only**
#### 1. Class

- A class is a blueprint or template used to create objects.

#### 2. Object

- An object is an instance of a class that represents a real-world entity.

#### 3. Attribute

- An attribute is a variable that belongs to a class or object.

#### 4. Method

- A method is a function defined inside a class.

#### 5. Encapsulation

- Encapsulation is the process of binding data and methods together into a single unit and restricting direct access to data.

#### 6. Abstraction

- Abstraction is the process of hiding implementation details and showing only the essential features of an object.

#### 7. Inheritance

- Inheritance is the mechanism by which one class acquires the properties and methods of another class.

#### 8. Polymorphism

- Polymorphism is the ability of the same method or interface to perform different actions based on the object.

#### 9. Constructor

- A constructor is a special method that is automatically executed when an object is created.

#### 10. Default Constructor

- A default constructor is a constructor that does not accept any parameters other than self.

#### 11. Parameterized Constructor

- A parameterized constructor is a constructor that accepts one or more parameters during object creation.

#### 12. Instance Variable

- An instance variable is a variable whose value is unique for each object of a class.

#### 13. Class Variable

- A class variable is a variable shared among all objects of a class.

#### 14. Instance Method

- An instance method is a method that operates on instance variables and requires an object to be called.

#### 15. Class Method

- A class method is a method that operates on class variables and uses the class itself as a parameter.

#### 16. Static Method

- A static method is a method that does not depend on instance variables or class variables.

#### 17. Method Overloading

- Method overloading is the ability to define multiple methods with the same name but different parameters.

#### 18. Method Overriding

- Method overriding is the process of redefining a method in a child class that already exists in the parent class.

#### 19. Abstract Class

- An abstract class is a class that contains one or more abstract methods and cannot be instantiated directly.

#### 20. Abstract Method

- An abstract method is a method declared without implementation that must be implemented by child classes.

#### 21. Interface

- An interface is a contract that specifies methods that a class must implement.

#### 22. Single Inheritance

- Single inheritance is the inheritance of one child class from one parent class.

#### 23. Multiple Inheritance

- Multiple inheritance is the inheritance of one child class from more than one parent class.

#### 24. Multilevel Inheritance

- Multilevel inheritance is a chain of inheritance where a class inherits from another derived class.

#### 25. Hierarchical Inheritance

- Hierarchical inheritance occurs when multiple child classes inherit from the same parent class.

#### 26. Hybrid Inheritance

- Hybrid inheritance is a combination of two or more types of inheritance.

#### 27. Data Hiding

- Data hiding is the practice of restricting direct access to an object's internal data.

#### 28. Access Specifier

- An access specifier is a mechanism that controls the visibility and accessibility of class members.

#### 29. Private Member

- A private member is accessible only within the class in which it is defined.

#### 30. Protected Member

- A protected member is accessible within the class and its derived classes.

#### 31. Public Member

- A public member is accessible from anywhere in the program.

#### 32. super()

- The super() function is used to access methods and constructors of a parent class.

#### 33. MRO (Method Resolution Order)

- MRO is the order in which Python searches for methods and attributes in an inheritance hierarchy.

#### 34. Association

- Association is a relationship where two independent classes are connected and can interact with each other.

#### 35. Aggregation

- Aggregation is a "has-a" relationship where one object contains another object, but both can exist independently.

#### 36. Composition

- Composition is a strong "has-a" relationship where the contained object cannot exist independently of the container object.

#### 37. Object-Oriented Programming (OOP)

- Object-Oriented Programming is a programming paradigm that organizes software design around objects, classes, and their interactions.
