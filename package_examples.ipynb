```python
#  wap ask the user take two number print subtraction multiplication and division
# a=eval(input())
# b=eval(input())
# c=a+b
import random
a=random.randint(1,100)
b=random.randint(1,100)
sub=a-b
mul=a*b
div=a/b
print(f"a:{a}")
print(f"b:{b}")
print('subtration:',sub)
print('multiplication:',mul)
print('division:',div)
```

```python
# wap ask the user take the radius find the area of circle
# pi=3.14
# r=100
# area_of_circle=pi*r*r
import random
import math
radius=random.randint(1,200)
area_of_circle=math.pi*radius*radius
print('radius=',radius)
print('area_of_circle=',area_of_circle)
```

```python
# wap ask the user take height and breadth find the area of traingle
#       area_of_traingle=0.5*b*h

import random
h=random.randint(1,50)
b=random.randint(1,50)
print("Height:",h)
print("Breadth:",b)
area_of_triangle=0.5*h*b
print("area_of_triangle is :",area_of_triangle)
```

```python
# wap ask the user take 3 numbers find the average
import random
a=random.randint(1,100)
b=random.randint(1,100)
c=random.randint(1,100)
print("num1:",a)
print("num2:",b)
print("num3:",c)
avg=(a+b+c)/3
print(f"The average of {a},{b},{c} is {avg}")
```

```python
# wap ask the user take the bill amount and tip anount calculate total bill
import random
bill_amount=eval(input("Enter bill amount :"))
tip_amount=random.randint(20,100)
total_bill=bill_amount+tip_amount
print("Your Total bill is:",total_bill)
```

```python
# wap ask the user take the salary, percetage of tax calculate tax amount
import random
salary=eval(input("Enter your Salary:"))
tax=random.randint(1,5)
print("Tax percentage is:",tax)
tax_amount=salary*tax/100
print("Tax amount is:",tax_amount)
```
